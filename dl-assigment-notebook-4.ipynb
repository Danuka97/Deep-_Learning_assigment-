# Task 1a

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
x_train = (x_train/255)
x_test = (x_test/255)
print(x_train.shape)
x_train= x_train.reshape(x_train.shape[0],(784))
x_test= x_test.reshape(x_test.shape[0],(784))
print(x_train.shape)
```

```python
#y_train = tf.one_hot(y_train, 10)
Y_train =tf.keras.utils.to_categorical(y_train, 10)
Y_test =tf.keras.utils.to_categorical(y_test, 10)
print(y_train.shape)
print(y_test.shape)
```

```python
import timeit

start = timeit.default_timer()
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(256,activation='relu'),
                             tf.keras.layers.Dense(128,activation='relu'),
                             tf.keras.layers.Dense(64,activation='relu'),
                             tf.keras.layers.Dense(10,activation='softmax')])

model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(),metrics=['acc'])
callbacks =[tf.keras.callbacks.EarlyStopping(
    monitor='val_acc', patience=3),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.1,
                              patience=2, min_lr=0.0001)]
end = timeit.default_timer()
print(end -start)
history = model.fit(x_train, y_train, verbose=1, validation_split=0.2, shuffle=True,callbacks=callbacks, epochs=30)
end = timeit.default_timer()
print(end - start)
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
import matplotlib.pyplot as plt

loss, accuracy  = model.evaluate(x_test, y_test, verbose=False)

plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='best')
plt.show()

print(f'Test loss: {loss:.3}')
print(f'Test accuracy: {accuracy:.3}')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='best')
plt.show()
```

# COV NN

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
x_train = (x_train/255)
x_test = (x_test/255)
X_train = x_train.reshape(x_train.shape[0], 28, 28,1)
X_test = x_test.reshape(x_test.shape[0], 28, 28,1)
print(X_train.shape)
print(X_test.shape)
Y_train =tf.keras.utils.to_categorical(y_train, 10)
Y_test =tf.keras.utils.to_categorical(y_test, 10)
print(y_train.shape)
print(y_test.shape)
```

```python
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D
```

```python
import timeit
start = timeit.default_timer()

model = tf.keras.Sequential([tf.keras.layers.Conv2D(32, (3,3), padding='valid',input_shape=(28,28,1)),
                             tf.keras.layers.MaxPooling2D(pool_size=2),
                             tf.keras.layers.Dropout(0.3),
                             tf.keras.layers.Conv2D(16, (2,2), padding='valid'),
                             tf.keras.layers.MaxPooling2D(pool_size=2),
                             tf.keras.layers.Flatten(),
                             tf.keras.layers.Dense(256,activation='relu'),
                             tf.keras.layers.Dense(128,activation='relu'),
                             tf.keras.layers.Dense(64,activation='relu'),
                             tf.keras.layers.Dense(10,activation='softmax')])

model.compile(optimizer='Adam', loss=tf.keras.losses.CategoricalCrossentropy(),metrics=['acc'])
callbacks =[tf.keras.callbacks.EarlyStopping(
    monitor='val_acc',verbose=2, patience=5),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.1,
                              patience=3,verbose=2, min_lr=0.0001)]

history = model.fit(X_train, Y_train, verbose=1, validation_split=0.2, shuffle=True,callbacks=callbacks, epochs=100)

end = timeit.default_timer()
print(end -start)
```

```python
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
import matplotlib.pyplot as plt

loss, accuracy  = model.evaluate(X_test, Y_test, verbose=False)

plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='best')
plt.show()

print(f'Test loss: {loss:.3}')
print(f'Test accuracy: {accuracy:.3}')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='best')
plt.show()
```

```python
import timeit
start = timeit.default_timer()

model = tf.keras.Sequential([tf.keras.layers.RandomContrast(0.3),
                             tf.keras.layers.Conv2D(32, (3,3), padding='valid',input_shape=(28,28,1)),
                             tf.keras.layers.MaxPooling2D(pool_size=2),
                             tf.keras.layers.Dropout(0.3),
                             tf.keras.layers.Conv2D(16, (2,2), padding='valid',activation = 'relu'),
                             tf.keras.layers.MaxPooling2D(pool_size=2),
                             tf.keras.layers.Flatten(),
                             tf.keras.layers.Dense(256,activation='relu'),
                             tf.keras.layers.Dense(128,activation='relu'),
                             tf.keras.layers.Dense(64,activation=tf.keras.layers.LeakyReLU(alpha=0.3)),
                             tf.keras.layers.Dense(10,activation='softmax')])

model.compile(optimizer='Adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(),metrics=['acc'])
callbacks =[tf.keras.callbacks.EarlyStopping(
    monitor='val_acc',verbose=2, patience=5,restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_acc', factor=0.1,
                              patience=2,verbose=2, min_lr=0.000001)]

history = model.fit(X_train, y_train, verbose=1, validation_split=0.2, shuffle=True,callbacks=callbacks, epochs=100)

end = timeit.default_timer()
print(end -start)
```

```python
import matplotlib.pyplot as plt

loss, accuracy  = model.evaluate(X_test, y_test, verbose=False)

plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='best')
plt.show()

print(f'Test loss: {loss:.3}')
print(f'Test accuracy: {accuracy:.3}')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='best')
plt.show()
```
